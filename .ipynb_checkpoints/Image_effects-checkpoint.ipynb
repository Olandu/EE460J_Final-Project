```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('surprised_pikachu.jpg')
plt.imshow(img)
plt.title('Original')
```

```python
# Blurring (works for both grayscale and color images)

def round_up_to_odd(f):
    f = int(np.ceil(f))
    return f + 1 if f % 2 == 0 else f

# Wrapper
def round_alpha(a):
    return round_up_to_odd(a*11)

def lpf_blur(img,alpha):
    n = round_alpha(alpha)
    # Blur (Averaging)
    blur = cv2.blur(img,(n,n))
    return blur

def gaussian_blur(img, alpha):
    n = round_alpha(alpha)
    # Gaussian Blur (passed through gauss filter)
    gauss_blur = cv2.GaussianBlur(img,(n,n), 0)
    return gauss_blur

def median_blur(img, alpha):
    n = round_alpha(alpha)
    # Median Blur
    median = cv2.medianBlur(img,n)
    return median
    
plt.figure()
plt.imshow(lpf_blur(img,0.7))
plt.title("LPF Blur")
plt.figure()
plt.imshow(gaussian_blur(img,0.7))
plt.title("Gaussian Blur")
plt.figure()
plt.imshow(median_blur(img,0.7))
plt.title("Median Blur")
```

```python
# Additive White Noise (works for both)

def additive_white_noise(img, alpha):
    h, w, c = img.shape

    mu = 0
    sigma = round(alpha*50)
    gauss_noise = np.random.normal(mu, sigma,size=(h,w,c))

    img_copy = img.copy()

    add_img = img_copy + gauss_noise
    neg_mask = add_img < 0
    add_img[neg_mask] = 0

    pos_mask = add_img > 255
    add_img[pos_mask] = 255
    add_img = add_img.astype(int)

    return add_img

plt.imshow(additive_white_noise(img, 0.3))
```

```python
# Salt and Pepper (works for both grayscale and color images)

def salt_and_pepper(img, alpha):
    h, w, c = img.shape

    thresh = round(alpha*0.25*127)

    noise = np.random.randint(255,size=(h,w,c))
    black = noise < thresh
    white = noise > 255 - thresh

    salt_and_pepper = img.copy()
    salt_and_pepper[white] = 255
    salt_and_pepper[black] = 0
    return salt_and_pepper

plt.imshow(salt_and_pepper(img, 0.8))
plt.title('Salt and Pepper')
```

```python
# Thesholding (only intended for grayscale images?)
# alpha value doesn't matter?

def threshold(img,alpha):   
    if img.shape[2] > 1:
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    else:
        img_gray = img
    ret, thresh = cv2.threshold(img_gray,127,255, cv2.THRESH_BINARY)
    return thresh

plt.imshow(threshold(img,0.2))
```

```python
# Zoom 

def zoom(img, alpha):
    h, w, _ = img.shape
    n = 1 + alpha

    zoom = cv2.resize(img, (0,0), fx=n, fy=n, interpolation=cv2.INTER_LINEAR)

    new_tl = np.array([0.5*(n-1)*h, 0.5*(n-1)*w]).astype(int)
    new_br = np.array([0.5*(n+1)*h, 0.5*(n+1)*w]).astype(int)
    
    zoom = zoom[new_tl[0]:new_br[0], new_tl[1]:new_br[1]]
    return zoom
```

```python
img = cv2.imread('surprised_pikachu.jpg')

plt.imshow(zoom(img, 0.1))
```

```python
def random_image_transform(img, alpha, show_img_flag=True, set_transform=None):
    
    if set_transform is None:
        n = np.random.randint(0,7)
    else:
        n = set_transform
        
    if n == 0:
        transform = lpf_blur(img, alpha)
        name = "Low Pass Filter Blur"
    elif n == 1:
        transform = gaussian_blur(img,alpha)
        name = "Gaussian Blur"
    elif n == 2:
        transform = median_blur(img, alpha)
        name = "Median Blur"
    elif n == 3:
        transform = additive_white_noise(img, alpha)
        name = "Additive White Noise"
    elif n == 4:
        transform = salt_and_pepper(img, alpha)
        name = "Salt and Pepper Noise"
    elif n == 5:
        transform = zoom(img, alpha)
        name = "Zoom"
    elif n == 6:
        transform = threshold(img, alpha)
        name = "Thresholding"
        
    if show_img_flag:
        plt.figure()
        plt.imshow(transform)
        plt.title(name)
    else:
        return transform

np.random.seed(69420)
for i in range(0,10):
    random_image_transform(img, 0.1)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import os
from os import listdir
from os.path import isfile, join, splitext
from IPython.display import Image, display

# Functions and classes for loading and using the Inception model.
import inception
```

```python
model = inception.Inception()
```

```python
def classify(image_path, index_flag=False, cls=None):
    # Display the image.
    # display(Image(image_path))

    # Use the Inception model to classify the image.
    pred = model.classify(image_path=image_path)

    pred = np.squeeze(pred)
    
    cls_source = np.argmax(pred)
    
    # Print the scores and names for the top-10 predictions.
    names, scores = model.print_scores(pred=pred, k=10, only_first_name=True)
    if cls is not None:
        cls_scores = []
        for cl in cls:
            cls_scores.append(pred[cl])
        return cls_scores
    if index_flag:
        return names, scores, cls_source
    else:
        return names, scores
```

```python
mypath = 'test_images/'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
onlyfiles.sort(key=len)
test_subset = [9, 60, 63, 82, 164, 199, 216, 350, 405, 494]
prediction_scores = {}
predictions = []
prediction_indexes = []

for i in test_subset:
    names,scores, index = classify(image_path=mypath+onlyfiles[i], index_flag=True)
    print(names[0], scores[0], index)
    prediction_scores[names[0]] = [scores[0]]
    predictions.append(names[0])
    prediction_indexes.append(index)
```

```python
transformations = ['lpf', 'gaussian', 'median', 'additive', 'salt', 'zoom']

# Initialize the dictionary for data
recorded_scores = {}
for i in range(len(test_subset)):
    img_number = test_subset[i]
    img_name = onlyfiles[img_number].split('.')[0]
    transform_dict = dict()
    for transform in transformations:
        original_score = prediction_scores[predictions[i]][0]
        transform_dict[transform] = [original_score]
    recorded_scores[img_name] = transform_dict

for i in range(len(test_subset)):
# for i in range(3,4):
    img_number = test_subset[i]
    img_name =  onlyfiles[img_number].split('.')[0]
    img = cv2.imread(mypath+onlyfiles[img_number])
    original_prediction = predictions[i]
    class_index = prediction_indexes[i]
    for j in range(len(transformations)):
        transform = transformations[j]
        for alpha in np.linspace(0.1,1,10):
            altered_img = random_image_transform(img,alpha, set_transform=j, show_img_flag=False)
            altered_img_name = 'altered_images/test_'+str(img_number)+'_'+transformations[j]+'_'+str(round(alpha,2))+'.jpg'
            cv2.imwrite(altered_img_name, altered_img)
            print(altered_img_name)
            
            altered_score = classify(image_path=altered_img_name, cls=[class_index])
            recorded_scores[img_name][transform].append(altered_score[0])
```
